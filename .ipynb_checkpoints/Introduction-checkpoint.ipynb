##### Experimental Method.
In this we implement the algorithm in a language and run it on different inputs.We record the exact running time in a unit like seconds microseconds.

##### Analytical method(Asymptotic analysis).
In this method we calculate running time of algorithm based on the input size.


### Asymptotic or Analytical analysis

Asymptotic analysis is the study of how therunning time of algorithm increases with the increase in input size.So in asymptotic analysis we dont care about the running time we are just interested in seeing what happens when input becomes large. 

Suppose we have two algorithms and there time when input size increases-

Input size    2     4    7    100    1000    

Algorithm1    4     8    14   200    2000

Algorithm2    4     16   49   10,000 1,000,000

We see that algorithm2 increases very rapidly with increase in the input size so algorithm 1 is better.

#### To find the rate of growth of running time we use -----> BIG O Notation.

#### Big O notation:
It shows us the asymptotic behaviour of a function -> how fast a function f(n) grows as n becomes larger.

#### Definitions

f is order of g.
f(n) is O(g(n)) if there exist constants c and n0(pronounce n not)
such that

f(n) <= g(n) for all n > n0 

In simple words function f will not go faster then function g for larger values of n.
f is the function whose rate of growth we have t examine and g is the function whose rate of growth we have to know.

Example-
suppose f(n) = 5n +4 and g(n) = n
By definitions the value of function f should always be less then or equal to the value of function g.
Let us assume the vlaue of constant c=6 and n0 = 4

now 5(4) + 4 <= 6 (4)
    24 <= 24
    
So we can say that f is order of g.

#### Finding Big O-
Suppose we have a function f and we have to find the order of the function-
f(n)=3n + 4n + 15
find f(n) is O(g(n))

suppose n = 10
f(n) = 300 + 40 + 15
Let us see the contribution of each value in the above expression.
Contribution of 3n:
3n forms 84.50% of total

Contribution of 4n
4n forms 11.26%

Contribution of 15
15 forms 4.22% of all.

If we increase the value of n the value of 3n will dominate the value of function and 4n and 15 will become negligible.
This is lso prvoed in above table as value of n is more fast growing then n or constant.

#### So when n becomes large Fastest growing term becomes dominent and lower terms,constants becomes insignificant.

So we can create the rules for find the big O-
* Keep the fastest growing term and discard rhe lower terms and constants.
* ignore coefficients.
  e.g f(n) = 5n we can say just f(n) is O(n) {ignore 5}
* If f(n) is constant, then we can say that f(n) is O(1)
* Base of algorithm is not important. 
    e.g. f(n) = 8logbase2n => f(n) is O(logn).
 

#### Tight and loose upper bounds.
For a function f(n) there can be many functions g(n) such that f(n) O(g(n))
Take an example-
f(n) =  5n + 4n + 8 is O(n)
but by definitions each of the function that has greater growing rate is also order of n
like O(n!), O(n) and all other.
So n is least upper bound or tight upper bound and all others that has greater growing rate is loose upper bounds.
##### our aim is to  always find tight upper bound.

#### Big O analysis of algorithm-
* Express the running time as a function of input size(n)
e.g T(n) = Running time in terms of n (time complexity of algorithm)
* Find big O for T(n)


#### Calculating time complexity-
Running time is proportional to the number of primitive operations executed during run time.
These primitive operations executed during run time can be comparisions,Arithmetic operations, Assignments, Input, output.
Now to find function T - Express the number of primitive operations executed during run time, in terms of n.
suppose we have algorithm-

```python
def SomeAlgorithm(int n):
    Input                                        c1
    Assignment                                   c2
    
    for i in range(n):                           c3
        Camparison
    
    Arithmetic operation.                        c4
```

Each of the operation takes time c1,c2,c3 and c4.Since comparision is inside a loop in will occur n times.

So total running time- 
T(n) = c1 + c2 + c3*n + c4
Lets find the O(n) for these functions-
T(n) is O(n)
#### So the definition  of asymptotic analysis we are  not  interested in how much time an operation takes to execute. We are interested in how many times an operation is performed.

### Big O analysis of different types of algorithm -
Case1-

```python
def Somealgo(int n):
    Input
    for i in range(n, 0, 4)
        operation
    Output
```

In above algo we see that the loop will run n/4 times(ignoring constant and lower power)
So T(n) = [n/4]
T(n) = O(n)
#### so we can say that time complexity of loop in which controlled variable is incremented or decremented by a constant amount is O(n)

Case2-
Let us see the case in which controlled variable is muliplied or devided by a constant in each oteration.

```python
def SomeAlgo(int n):
    Assignment
    i = n-1
    while i>=1:
        Arithmetic operation
        i/2=2
    
    Output
```

In above code if n=51 is passed it will run 6 times.
so T(n) = [log2n]
We know in big O base is ignored.
so T(n) = O(log n)
#### Now let us multiply the constant in each iteration.

```python
def SomeAlgo(int n):
    Assignment
    i = n-1
    while i>=1:
        Arithmetic operation
        i*=2
    
    Output
```

If we pass n= 51 it will still run 6 times. so T(n) = O(log n)
#### So if we multiply or devide the controlled variable with a constant the time complexity will remain same.

Case3- Nested Loops-

```python
def SomeAlgo(int n):
    Input
    
    for i in range(n):
        Assignment
        for in range(n):
            Comparision
            
    Arithmetic operation
    Output
    
```

So in nested loop outer loop will run n times and inner loop will also run n* n times.
Total iteration = n + n*n
T(n) = n + n
T(n) is O(n)
So number of nested loops increses the power if the nested has k inside j it will become O(n).

### We can have three case for an algo anlysis in  finding time complexity-

#### Worst 
* Maximum running time that an algorithm will take for input size n.
* Considers an input for which the algorithm takes the longest time. 
* e.g suppose we want find an elemnt in a linear search and the element is present at last position.
#### Best case
* minimum running time that an algorithm will take for input size n.
* Considers an input for which the algorithm takes the shortest time.
* e.g suppose we want find an element in a linear search and the element is present at first position.
#### Average
* average running time that an algorithm will take for input size n.
* consider all possible inputs and compute average.

### In practice Worst case analysis is usually done to analyse algorithm becaue in this we get to know the most time taken by an algo to run. Actual running time can never be worst then this.
